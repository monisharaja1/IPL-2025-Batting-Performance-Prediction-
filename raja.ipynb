```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

# Load the dataframe again as the previous cells were not executed
df = pd.read_csv('IPL2025Batters.csv')

# Data Cleaning and Transformation
df['HS'] = df['HS'].str.replace('*', '', regex=False)
df['AVG'] = df['AVG'].replace('-', '0')
df['HS'] = pd.to_numeric(df['HS'])
df['AVG'] = pd.to_numeric(df['AVG'])

# One-hot encode the 'Team' column
df = pd.get_dummies(df, columns=['Team'], drop_first=True)

scaler = StandardScaler()
list2=['Runs','Matches','Inn','HS','BF','4s','6s', 'AVG' ]
df[list2] = scaler.fit_transform(df[list2])
df.head()
```

```python
df.dtypes
```

```python
core = df.drop(['Player Name', 'Team'], axis=1).corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(core, annot=True, cmap='plasma')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Define x and y after one-hot encoding
x = df.drop(['Player Name', 'Runs'], axis=1)
y = df['Runs']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
model_accuracy=model.score(x_test,y_test)*100
model_accuracy
```

```python
plt.scatter(y_test,y_pred)
plt.xlabel('Actual Runs')
plt.ylabel('Predicted Runs')
plt.title('Actual vs Predicted Runs')
plt.show()
```

```python
predicted=y_test-y_pred
plt.scatter(y_test,predicted)
plt.xlabel('Actual Runs')
plt.ylabel('Prdicted Runs')
plt.title('Actual vs Predicted')
plt.show()
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
```
